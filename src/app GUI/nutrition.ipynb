```python
import pandas as pd
import numpy as np
```

# Importing the Data

1. `recipes`  : the primary purpose of this data is to get the ID of the recipe/food item and its associated ingredients
2. `nutrients`: the primary purpose of this data is to get the ID of the recipe/food item and its associated nutrition values

The datasets will be join on the `Food ID` column in order to get the ID of the recipe, their ingredients and nutrition values.

```python
# getting the recipes data
recipes = pd.read_csv("/Users/shreeshagarwal/Desktop/PIC16/recipe-app/Recipes.csv")
print(recipes.shape)
recipes.head()
```

```python
# getting the nutrition data
nutrients = pd.read_excel("/Users/shreeshagarwal/Desktop/PIC16/recipe-app/nutrients.xls")
print(nutrients.shape)
nutrients.head()
```

# Removing Duplicates From Recipes

```python
recipes =recipes.drop_duplicates(subset=['Food ID'])
print(recipes.shape)
recipes.head()
```

# Combining the Datasets

```python
food_info = pd.merge(recipes, nutrients, on= "Food ID")
print(food_info.shape)
food_info.head()
```

# Cleaning the Data

```python
df = food_info.copy()
# dropping redundant columns
df = df.drop(["Food Survey ID","Survey flag_x",'Total Weight Change (%)','Survey ID', 'Food Name_y', 'Survey flag_y', 'Ingredient Survey ID', 'Retention Factor',
             'Ingredient Weight (g)', 'C20:5w3 Eicosapentaenoic (mg)', 'C22:6w3 Docosahexaenoic (mg)', 'C22:5w3 Docosapentaenoic (mg)',
             'Tryptophan (mg)', 'Ingredient ID'], axis = 1) 
# setting index to Food ID
df.set_index(keys=["Food ID"])
df.head()
```
